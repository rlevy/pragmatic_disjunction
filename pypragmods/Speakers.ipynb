# The speaker perspective

```python
import itertools
import cPickle as pickle
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import lsa
```

```python
%matplotlib inline
```

## Small lexicon

```python
experiment = lsa.Experiment(n=10, 
                            baselexicon={'A': ['1'], 'B': ['2'], 'X':['1', '2']}, 
                            lexical_costs={'A':0.0, 'B':0.0, 'X':0.0},
                            temperature=1.0, 
                            disjunction_cost=1.0, 
                            beta=1.0, 
                            alpha=2.0, 
                            prior=None,    # Triggers a flat prior.
                            lexprior=None, # Triggers a flat prior.
                            null_cost=5.0)
experiment.build()
```

### Listener inferences for the above settings

```python
experiment.display_listener_inference(msg='A v X')
```

### Speaker table for P(msg|Lex) for the above settings

```python
experiment.display_speaker_behavior(state='1 v 2')
```

### Disjunction costs: Definitional intentions

```python
experiment.plot_speaker_behavior_disjunction_costs(state='1', lexicon=0, progress_report=False)
```

### Disjunction costs: Hurfordian lexicon and disjunctive state

```python
experiment.plot_speaker_behavior_disjunction_costs(state='1 v 2', lexicon=1, progress_report=False)
```

### Disjunction costs: Anti-Hurfordian lexicon and disjunctive state

```python
experiment.plot_speaker_behavior_disjunction_costs(state='1 v 2', lexicon=2, progress_report=False)
```

## Large lexicon

```python
large = lsa.Experiment(n=10,
                       baselexicon={'A': ['1'], 'B': ['2'], 'C': ['3'], 'D': ['4'], 'X': ['1', '2', '3', '4']}, 
                       lexical_costs={'A':0.0, 'B':0.0, 'C':0.0, 'D':0.0, 'X':0.0},
                       temperature=1.0, 
                       disjunction_cost=0.00,
                       beta=12.0, 
                       alpha=8.0, 
                       prior=None,    # Triggers a flat prior.
                       lexprior=None, # Triggers a flat prior.
                       null_cost=5.0)
large.build()
```

### Best listener inferences for the above setting

```python
lis_prob_table = large.listener_inference(msg='A v X')
large.show_max_lex_state_values(lis_prob_table)
```

### Best speaker behavior for the above settings: definitional target state

```python
spk_prob_table = large.speaker_behavior(state='1')
large.show_max_message_values(spk_prob_table)
```

### Best speaker behavior for the above settings: disjunctive target state

```python
spk_prob_table = large.speaker_behavior(state='1 v 2')
large.show_max_message_values(spk_prob_table)
```

### Disjunction costs: definitional intentions

```python
large.plot_speaker_behavior_disjunction_costs(state='1', lexicon=0, progress_report=False, 
                                              disjunction_cost_values=np.arange(0.0, 0.5, 0.1))
```

### Disjunction costs: Hurfordian lexicon and disjunctive state

```python
large.plot_speaker_behavior_disjunction_costs(state='1 v 2', lexicon=1, progress_report=False, 
                                              disjunction_cost_values=np.arange(0.0, 0.5, 0.1))
```

### Disjunction costs: Anti-Hurfordian lexicon and disjunctive state

```python
large.plot_speaker_behavior_disjunction_costs(state='1 v 2', lexicon=4, progress_report=False, 
                                              disjunction_cost_values=np.arange(0.0, 0.5, 0.1))
```
