# Understanding definitional readings

#### The notebook seeks to understand definitional readings via parameter exploration.

```python
import itertools
import cPickle as pickle
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import lsa
```

```python
%matplotlib inline
```

## 1. Create data sets for lexica with 3, 4, and 5 atomic lexical items

Note: this takes a LOOOONG time; the results are pickled and can be read in with the command below.
I'm including this code here mainly to show the space explored. Also, once we have a clearer picture of the exact model we
want to advocate for, we can use Parallel Python to explore much larger parameter spaces efficiently.

```python
# lsa.explore_hyperparameters creates a mapping from (lexindex, state) pairs to parameters that support them. 
# The parameters are given as dicts in this form:
#
#  {'lambda': temp, 'alpha': alpha, 'beta': beta, 'depth': depth, 'disjunction_cost': dcost, 'prob': float}
#
# where prob is the probability, included because sometimes low values indicate a lack of stability.

lexica = [{'A': ['1'], 'B': ['2'], 'X':['1', '2']},
          {'A': ['1'], 'B': ['2'], 'C': ['3'], 'X': ['1', '2', '3']},
          {'A': ['1'], 'B': ['2'], 'C': ['3'], 'D': ['4'], 'X': ['1', '2', '3', '4']}]

for lex in lexica:
    lexical_costs = {let:0.0 for let in lex.keys()}    
    results = lsa.explore_hyperparameters(baselexicon=lex, 
                                          lexical_costs=lexical_costs, 
                                          msg='A v X',
                                          temps=[1.0], # Fixed at 1 to trivialize it.
                                          dcosts=np.arange(0.0, 0.21, 0.01), # We know definitional readings require low costs.
                                          alphas=np.arange(0.0, 16.0, 1), # Give alpha and beta lots of room -- required
                                          betas=np.arange(0.0, 16.0, 1),  # if lambda is fixed at 1.0.
                                          depths=[10]) # Iterate to depth 10 to try to ensure stability.
    output_filename = 'paramexplore-defin-lex%s.pickle' % len(lex)
    pickle.dump(results, file(output_filename, 'w'), 2)
```

```python
# Load the pre-computed parameter values:

paramexplore_lex3 = pickle.load(file('paramexplore-defin-lex3.pickle'))
paramexplore_lex4 = pickle.load(file('paramexplore-defin-lex4.pickle'))
paramexplore_lex5 = pickle.load(file('paramexplore-defin-lex5.pickle'))
```

```python
# Define this for convenience. It will always be our target for this notebook:

definitional_lexindex_state_pair = (0, '1')
defin_lex3 = paramexplore_lex3[definitional_lexindex_state_pair]
defin_lex4 = paramexplore_lex4[definitional_lexindex_state_pair]
defin_lex5 = paramexplore_lex5[definitional_lexindex_state_pair]
```

## 2. Look at the min and max values for the parameters

Reminder: given our exploration, alpha and beta are bounded by [0,14] and disjunction_cost is bounded by [0,0.2]

```python
def inspect_params(vals):
    for param in ('alpha', 'beta', 'disjunction_cost'):
        param_vals = [r[param] for r in vals]
        print "\t%s: [%s, %s]" % (param.rjust(18), min(param_vals), max(param_vals))
```

### The 3-item lexicon

```python
inspect_params(defin_lex3)
```

### The 4-item lexicon

```python
inspect_params(defin_lex4)
```

### The 5-item lexicon

```python
inspect_params(defin_lex5)
```

### Generalizations:

* It seems that larger lexicons demand smaller disjunction costs.
* It seems that larger lexicons demand larger alpha and beta. 

## 3. Pairwise scatterplots

```python
def param_scatterplots(vals):
    """Plot the pairwise relationships between our parameters for some reading (<Lex, state> pair) of interest."""
    fig, axarray = plt.subplots(nrows=1, ncols=3)
    fig.set_figheight(10)
    fig.set_figwidth(30)
    params = ('alpha', 'beta', 'disjunction_cost')
    ax_index = 0    
    for i in range(len(params)-1):
        p1 = params[i]
        for j in range((i+1), len(params)):
            p2 = params[j]
            if p1 != p2:
                twoparam_scatterplot(axarray[ax_index], vals, p1, p2) 
                ax_index +=1
            
def twoparam_scatterplot(ax, vals, p1, p2):
    vals1 = [r[p1] for r in vals]
    vals2 = [r[p2] for r in vals]
    ax.plot(vals1, vals2, linestyle="", marker=".")
    ax.set_xlabel(p1)
    ax.set_ylabel(p2)
```

### The 3-item lexicon

```python
param_scatterplots(defin_lex3)
```

### The 4-item lexicon

```python
param_scatterplots(defin_lex4)
```

### The 5-item lexicon

```python
param_scatterplots(defin_lex5)
```

### Generalizations:

1. Definitional readings *almost* require beta > alpha.
2. Higher beta allows for higher disjunction costs.
3. High and low alpha require smaller disjunction costs.
4. Points 2 and 3 likely trace to the relationship between alpha and beta implied by point 1.

## 4. The beta/alpha ratio vs. disjunction costs

The above suggests a sysetmatic dependency between alpha and beta. Here, I compare their ratio to the disjunction costs for each of the three lexicon sizes.

```python
def alpha_beta_gamma_scatterplot():
    fig, axarray = plt.subplots(nrows=1, ncols=3)
    fig.set_figheight(10)
    fig.set_figwidth(30)
    for i, df in enumerate((defin_lex3, defin_lex4, defin_lex4)):
        alpha_beta = [r['beta']/r['alpha'] for r in df]
        disj = [r['disjunction_cost'] for r in df]
        axarray[i].plot(alpha_beta, disj, linestyle="", marker=".")
        axarray[i].set_xlabel('beta/alpha')
        axarray[i].set_ylabel('disjunction_cost')   
        axarray[i].set_title("Lexicon size: %s" % (i+3)) 
```

```python
alpha_beta_gamma_scatterplot()
```
