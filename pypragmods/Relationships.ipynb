# Understanding the relationship between definitional and Hurfordian readings

## 1. Load data sets for lexica with 3, 4, and 5 atomic lexical items

```python
# Load the pre-computed parameter values:
paramexplore_lex3 = pickle.load(file('paramexplore-defin-lex3.pickle'))
paramexplore_lex4 = pickle.load(file('paramexplore-defin-lex4.pickle'))
paramexplore_lex5 = pickle.load(file('paramexplore-defin-lex5.pickle'))
```

```python
# 3-item, 4-item, and 5-item base lexica:
base_lexica = [{'A': ['1'], 'B': ['2'], 'X':['1', '2']},
               {'A': ['1'], 'B': ['2'], 'C': ['3'], 'X': ['1', '2', '3']},
               {'A': ['1'], 'B': ['2'], 'C': ['3'], 'D': ['4'], 'X': ['1', '2', '3', '4']}]
```

```python
# Relative to our utterance 'A v X':
# - a Hurfordian lexicon is one where A and X have disjoint denotations.
# - a definitional lexicon is one where A == X
def classify_lexica(base_lexicon):
    classified = []
    experiment = lsa.Experiment(baselexicon=base_lexicon, lexical_costs={key:0.0 for key in base_lexicon.keys()})
    experiment.build()
    for i, lex in enumerate(experiment.lexica):
        a_index = experiment.messages.index('A')
        a_sem = set([s for j, s in enumerate(experiment.states) if lex[a_index][j] > 0.0])
        x_index = experiment.messages.index('X')
        x_sem = set([s for j, s in enumerate(experiment.states) if lex[x_index][j] > 0.0])
        cls = 'O'
        if len(a_sem & x_sem) == 0:
            cls = 'H'            
        elif a_sem == x_sem:
            cls = 'D'
        classified.append((i, experiment.lex2str(i), cls))                              
    return classified                        
```

```python
# View the lexica to check assumptions:
def view_lex_classes(cls):
    for base_lexicon in base_lexica:
        print "=" * 70
        print base_lexicon
        for i, lex, c in classify_lexica(base_lexicon):        
            if c == cls:
                print i, lex
```

```python
view_lex_classes('H')
```

```python
# Cluster parameters by lexicon and class:
def get_cls_params(base_lexicon, cls, state, paramexplore_lex):
    cls_lex = []
    for lexindex, lex, c in classify_lexica(base_lexicon):
        if c == cls:
            cls_lex += paramexplore_lex[(lexindex, state)]
    return cls_lex
```

```python
hc_lex3 = get_cls_params(base_lexica[0], 'H', '1 v 2', paramexplore_lex3)
hc_lex4 = get_cls_params(base_lexica[1], 'H', '1 v 2', paramexplore_lex4)
hc_lex5 = get_cls_params(base_lexica[2], 'H', '1 v 2', paramexplore_lex5)

defin_lex3 = get_cls_params(base_lexica[0], 'D', '1', paramexplore_lex3)
defin_lex4 = get_cls_params(base_lexica[1], 'D', '1', paramexplore_lex4)
defin_lex5 = get_cls_params(base_lexica[2], 'D', '1', paramexplore_lex5)
```

## Plotting the relationship between log(beta/alpha) and gamma (disjunction cost)

```python
colors = ['#1B9E77', '#D95F02', '#7570B3', '#E7298A', '#66A61E', '#E6AB02', '#A6761D', '#666666']

param_pairs = [(hc_lex3, defin_lex3),
               (hc_lex4, defin_lex4),
               (hc_lex5, defin_lex5)]

def alpha_beta_gamma_scatterplot(param_pairs, logx=True):
    transform = np.log if logx else (lambda x : x)    
    fig, axarray = plt.subplots(nrows=3, ncols=1)
    fig.set_figheight(30)
    fig.set_figwidth(20)    
    for i, params in enumerate(param_pairs):
        hc, defin = params
        ax = axarray[i]
        # Hurfordian:
        h_ba = [r['beta']/r['alpha'] for r in hc]
        h_g = [r['disjunction_cost'] for r in hc]
        # Definitional:
        d_ba = [r['beta']/r['alpha'] for r in defin]
        d_g = [r['disjunction_cost'] for r in defin]        
        # Both (if any):
        both = [(x,y) for x, y in zip(d_ba, d_g) if (x,y) in zip(h_ba, h_g)]
        b_ba, b_g = zip(*both)
        # Remove overlap:
        h_ba, h_g = zip(*[(x,y) for x,y in zip(h_ba, h_g) if (x,y) not in both])
        d_ba, d_g = zip(*[(x,y) for x,y in zip(d_ba, d_g) if (x,y) not in both])                        
        # Plotting:
        for i, vals in enumerate(((h_ba, h_g, 'Hurfordian'), (d_ba, d_g, 'Definitional'), (b_ba, b_g, 'Both'))):
            x, y, label = vals
            ax.plot(jitter(transform(x)), jitter(y), linestyle="", marker=".", label=label, markersize=8, color=colors[i])
        # In case we want to specify the axis limits:
        x1,x2,y1,y2 = ax.axis()
        ax.axis((x1, x2, y1, y2))
        # Labeling:
        if logx:
            ax.set_xlabel(r'$\log(\beta/\alpha)$')
        else:
            ax.set_xlabel(r'$\beta/\alpha$')
        ax.set_ylabel('disjunction_cost')   
        ax.set_title("Lexicon size: %s" % (i+3)) 
        ax.legend(bbox_to_anchor=(1.1, 1.1))
        
def jitter(x):
    """Jitter while respecting the data bounds"""
    mu = 0.0
    sd = 0.001
    j = x + np.random.normal(mu, sd, len(x))
    j = np.maximum(j, np.min(x))
    j = np.minimum(j, np.max(x))
    return j
    
alpha_beta_gamma_scatterplot(param_pairs, logx=True)
```

## Histograms of the beta/alpha values for each reading

```python
def beta_alpha_histogram(hc=None, defin=None, lexicon_size=None, logx=True, xmin=-3.0, xmax=3.0):
    transform = np.log if logx else (lambda x : x) 
    fig, axarray = plt.subplots(nrows=1, ncols=2)
    fig.set_figheight(10)
    fig.set_figwidth(20)
    if logx:
        xlabel = r'$\log(\beta/\alpha)$'
    else:
        xlabel = r'$\beta/\alpha$'
    ylabel = "Density"
    for i, vals in enumerate(((hc, 'Hurfordian'), (defin, 'Definitional'))):
        x, label = vals
        ba = [(r['beta']/r['alpha'], (r['beta'], r['alpha'])) for r in x]
        ba_sorted = sorted(ba)
        ba, ba_pairs = zip(*ba_sorted)
        axarray[i].hist(transform(ba), color=colors[i], normed=True)
        axarray[i].set_xlabel(xlabel)
        axarray[i].set_ylabel(ylabel)
        x1, x2, y1, y2 = axarray[1].axis()
        axarray[i].axis((xmin, xmax, y1, y2))
        min_ba_str = "%s/%s" % ba_sorted[0][1]
        min_ba_str += " = %s" % np.round(ba_sorted[0][0], 3)
        title = "%s, lexicon size: %s; min(beta/alpha) = %s" % (label, lexicon_size, min_ba_str)
        axarray[i].set_title(title)
```

```python
beta_alpha_histogram(hc=hc_lex3, defin=defin_lex3, lexicon_size=3)
```
