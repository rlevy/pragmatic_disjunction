### The basic set-up

We'll be keeping all of the parameters fixed except for disjunction_cost, which we will vary. The value specified here is
therefore arbitrary.

```python
experiment = lsa.Experiment(n=10, 
                            baselexicon={'A': ['1'], 'B': ['2'], 'X':['1', '2']}, 
                            lexical_costs={'A':0.0, 'B':0.0, 'X':0.0},
                            temperature=1.0, 
                            disjunction_cost=0.0, # Arbitrary; we'll be quantifying over it.
                            beta=1.0, 
                            alpha=2.0, 
                            prior=None,    # Triggers a flat prior.
                            lexprior=None, # Triggers a flat prior.
                            null_cost=5.0)
```

# Hello

### Definitional target state

```python
experiment.plot_listener_inference_disjunction_costs(msg='A v X', 
                                                     target_state='1', 
                                                     progress_report=False, 
                                                     disjunction_cost_values=np.arange(0.0, 0.3, 0.005))
```

### Hurfordian target state

```python
experiment.plot_listener_inference_disjunction_costs(msg='A v X',
                                                     target_state='1 v 2',
                                                     progress_report=False, 
                                                     disjunction_cost_values=np.arange(0.0, 0.3, 0.005))
```

### Anti-Hurfordian reading

If we set beta=0.0, then no lexical inferences are made and the anti-Hufordian lexicon wins out with modest disjunction costs.

```python
experiment2 = lsa.Experiment(n=10, 
                             baselexicon={'A': ['1'], 'B': ['2'], 'X':['1', '2']},  
                             lexical_costs={'A':0.0, 'B':0.0, 'X':0.0},
                             temperature=1.0, 
                             disjunction_cost=0.00, # Arbitrary; we'll be quantifying over it.
                             beta=0.0, 
                             alpha=2.0, 
                             prior=None,    # Triggers a flat prior.
                             lexprior=None, # Triggers a flat prior.
                             null_cost=5.0)

experiment2.plot_listener_inference_disjunction_costs(msg='A v X', 
                                                     target_state='1 v 2', 
                                                     progress_report=False, 
                                                     disjunction_cost_values=np.arange(0.0, 0.3, 0.005))
```
